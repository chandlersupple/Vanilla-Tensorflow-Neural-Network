```
import tensorflow as tf
sess = tf.InteractiveSession()
```

```
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot= True)
```

```
x = tf.placeholder(tf.float32, shape= [None, 784])
y = tf.placeholder(tf.float32, shape= [None, 10])
```

```
W1 = tf.Variable(tf.truncated_normal(shape= [784, 512]))
b1 = tf.Variable(tf.truncated_normal(shape= [512]))
W2 = tf.Variable(tf.truncated_normal(shape= [512, 10]))
b2 = tf.Variable(tf.truncated_normal(shape= [10]))
```

```
output1 = tf.nn.relu(tf.matmul(x, W1) + b1)
dropout1 = tf.nn.dropout(output1, keep_prob= 0.5)
output2 = tf.matmul(dropout1, W2) + b2

cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels= y, logits= output2))
```

```
optimizer = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy)
```

```
correct = tf.equal(tf.argmax(output2, 1), tf.argmax(y, 1))
accuracy = tf.reduce_mean(tf.cast(correct, dtype= tf.float32))
```

```
sess.run(tf.global_variables_initializer())
```

```
for epoch in range (100000):
  batch = mnist.train.next_batch(50)
  optimizer.run(feed_dict= {x: batch[0], y: batch[1]})
  if epoch % 1000 == 0:
    batch_acc = accuracy.eval(feed_dict= {x: batch[0], y: batch[1]})
    print("Step %s, Accuracy %s" %(epoch, batch_acc))
```

```
print(accuracy.eval(feed_dict= {x: mnist.test.images, y: mnist.test.labels}))
```
